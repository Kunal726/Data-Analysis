## Exploratory analysis and visualization

```python
# importing libraries
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_ds.Country.count()
```

```python
# Unique Countries
covid_ds.Country.unique()
```

## Number of cases with respect to region

```python
no_cases_region = covid_ds.groupby('Region')[['Total cases']].sum()

no_cases_region = no_cases_region.sort_values('Region', ascending = False).reset_index()

no_cases_region.rename(columns = {'Total cases' : 'total_cases'}, inplace = True)
```

```python
no_cases_region
```

```python
mp.rcParams['font.size'] = 14
mp.rcParams['figure.figsize'] = (15, 7.5)
mp.rcParams['figure.facecolor'] = '#00000000'
sns.set_style("darkgrid")
```

```python
plt.title("Cases with respect to Region")
plt.xlabel('Regions')
plt.ylabel('Total cases')
plt.plot(no_cases_region.Region, no_cases_region.total_cases);
```

## Number of Cases with respect to country

```python
no_cases_country = covid_ds.groupby('Country')[['Total cases']].sum()

no_cases_country = no_cases_country.sort_values('Country', ascending = True).reset_index()

no_cases_country.rename(columns = {'Total cases' : 'total_cases'}, inplace = True)
```

```python
plt.figure(figsize=(20, 85))
plt.title('Number of cases with respect to Country',fontweight=800)
sns.barplot(x='total_cases', y='Country', data=no_cases_country);
```

## Total deaths with respect to country

```python
covid_ds.columns
```

```python
no_deaths_country = covid_ds.groupby('Country')[['Total Deaths']].sum()

no_deaths_country = no_deaths_country.sort_values('Country', ascending = True).reset_index()

no_deaths_country.rename(columns = {'Total Deaths' : 'total_deaths'}, inplace = True)
```

```python
no_deaths_country
```

```python
plt.figure(figsize=(20, 85))
plt.title('Number of deaths with respect to Country',fontweight=800)
sns.barplot(x='total_deaths', y='Country', data=no_deaths_country);
```

## Total deaths with respect to Region

```python
no_deaths_region = covid_ds.groupby('Region')[['Total Deaths']].sum()

no_deaths_region = no_deaths_region.sort_values('Region', ascending = True).reset_index()

no_deaths_region.rename(columns = {'Total Deaths' : 'total_deaths'}, inplace = True)

no_deaths_region

plt.figure(figsize=(35, 15))
plt.title('Number of deaths with respect to Country',fontweight=800)
sns.barplot(y='total_deaths', x='Region', data=no_deaths_region);
```

# How many cases are there in India in last 7 days?

```python
cases_india = covid_ds[covid_ds.Country == 'India']
```

```python
cases_last_7_days = cases_india['Reported last7 days']
```

```python
print("Total cases in India in last 7 Days are {}".format(cases_last_7_days[1]))
```

# Total cases in Argentina?

```python
cases_Argentina = covid_ds[covid_ds.Country == 'Argentina']
```

```python
cases_total = cases_Argentina['Total cases'].sum()
```

```python
print("Total cases in Argentina are {}".format(cases_total))
```

# How many cases are there in India in total per Million?

```python
cases_india_m = covid_ds[covid_ds.Country == 'India']
cases_per_m = cases_india_m['Total Cases per 100000'].sum()
print("Total cases in India per Million are {} million".format(cases_per_m))
```

# Total cases in Africa Region?

```python
cases_africa = covid_ds[covid_ds.Region == 'Africa']
cases_africa = cases_africa['Total cases'].sum()
print("Total cases in Africa region are {} ".format(cases_africa))
```

# Which region has highest number of cases?

```python
cases_region = covid_ds.groupby('Region')[['Total cases']].sum()
cases_region.sort_values('Total cases', ascending=False)
```
